```python
import pandas as pd
import plotly
import plotly.graph_objects as go
import plotly.offline as offline

offline.init_notebook_mode(connected = True)
```

```python
dow_jones_data = pd.read_csv("datasets/DJI.csv")

dow_jones_data.head(5)
```

```python
dow_jones_data.shape
```

```python
trace_open = go.Scatter(x = dow_jones_data.Date,
                        y = dow_jones_data.Open,
                        name = "Stock Open",
                        line = dict(color = "red"))

data = [trace_open]

offline.iplot(data)
```

```python
trace_close = go.Scatter(x = dow_jones_data.Date,
                        y = dow_jones_data.Close,
                        name = "Stock Close",
                        line = dict(color = "blue"))

data = [trace_close]

offline.iplot(data)
```

```python
data = [trace_close, trace_open]

layout = dict(title = "Stock Data",
              
              xaxis = dict(rangeslider = dict(visible = True),
                           type = "date"))
```

```python
fig = dict(data = data,
           layout = layout)

offline.iplot(fig)
```
